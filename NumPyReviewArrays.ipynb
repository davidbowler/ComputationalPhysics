```python
n=101
t = np.linspace(0,2*np.pi,n)
#pos_vec = np.array((np.cos(t),np.sin(t)))
pos_vec = np.zeros((n,2))
pos_vec[:,0] = np.cos(t)
pos_vec[:,1] = np.sin(t)
```

```python
half_n = int(n/2)
plt.plot(t[0:half_n],pos_vec[0:half_n,0])
```

```python
step = 10
plt.plot(pos_vec[::step,0],pos_vec[::step,1])
plt.axis('scaled')
```

```python
plt.plot(t,np.roll(pos_vec[:,0],-10))
```
