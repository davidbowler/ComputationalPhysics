```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cubic_for_roots(x):
    """Implements x**3 - 2*x**2 -x + 2
    
    Input:  x
    Output: f(x)
    """
    return x**3 - 2*x**2 - x + 2
```

```python
x_plot = np.linspace(-2,3,101)
plt.plot(x_plot,cubic_for_roots(x_plot))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Example of root finding function")
plt.grid()
```

```python
# Set initial brackets by inspection
a = -2
b = -0.5
fa = cubic_for_roots(a)
fb = cubic_for_roots(b)
plt.plot(x_plot,cubic_for_roots(x_plot))
plt.plot(a,fa,'ro')
plt.plot(b,fb,'ro')
plt.text(a,fa,r'a$_0$')
plt.text(b,fb,r'b$_0$')
plt.grid()

# Set up iterations
n = 0
maxiters = 500
tol  = 1e-6
fc = 10*tol
while abs(fc)>tol and n<maxiters:
    # Mid-point of a and b
    c = (a+b)/2 
    fc= cubic_for_roots(c)
    # Test to see whether we replace a or b with c
    if fa*fc>0: 
        plt.plot(a,fa,'ko')
        a = c
        fa = fc
        plt.plot(a,fa,'ro')
        plt.text(a,fa,f'a{n+1}')
    else: 
        plt.plot(b,fb,'ko')
        b = c
        fb = fc
        plt.plot(b,fb,'ro')
        plt.text(b,fb,f'b{n+1}')
    n = n+1
print(f"Done {n} iterations")
print(f"f(c) is {fc:9.5e}")
print(f"c is {c:9.5e}")
```
